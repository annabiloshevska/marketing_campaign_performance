```python
import pandas as pd
import numpy as np
```

```python
#Loading a dataset
url = 'https://raw.githubusercontent.com/annabiloshevska/marketing_campaign_performance/refs/heads/master/data/raw/marketing_campaign_dataset.csv'
df = pd.read_csv(url)

display(df.head())
print(df.info())
display(df.describe())
display(df.describe(include='object'))
display(df.isnull().sum())
print(df.duplicated().sum())
```

```python
print(df.columns)
```

```python
df["Acquisition_Cost"] = (
    df["Acquisition_Cost"]
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
)

df["Acquisition_Cost"] = pd.to_numeric(df["Acquisition_Cost"], errors="coerce")
```

```python
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
```

```python
df["Duration_num"] = df["Duration"].str.extract(r"(\d+)").astype(float)
```

```python
# Creating core metrics

# CTR: Click-Through Rate
df["CTR"] = df["Clicks"] / df["Impressions"]

# CPC: Cost Per Click
df["CPC"] = df["Acquisition_Cost"] / df["Clicks"]

# CPM: Cost Per 1000 Impressions
df["CPM"] = (df["Acquisition_Cost"] / df["Impressions"]) * 1000

df.head()
display(df.info())
```

```python
df.to_csv("marketing_campaign_clean.csv", index=False)
```
